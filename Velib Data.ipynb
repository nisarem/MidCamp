```python
import requests
import pandas as pd
import time
from datetime import datetime
```

```python
def collect_data(velibs):
    try:
        response = requests.get("https://opendata.paris.fr/api/explore/v2.1/catalog/datasets/velib-disponibilite-en-temps-reel/exports/csv?lang=fr&timezone=Europe%2FBerlin&use_labels=true&delimiter=%3B")
        txt = response.text
        with open('velib.csv', 'w+') as f: #Question: where is this file stored currently? 
            f.write(txt)
            
        data = pd.read_csv("velib.csv", sep=";")
        data.columns = data.columns.str.lower().str.replace(' ', '_')
        data['update_time'] = pd.Timestamp.now()
        
        if velibs is None:
            velibs = data
        else:
            velibs = pd.concat([velibs, data], ignore_index=True)
            
        print("Data gathered successfully at:", pd.Timestamp.now())
        return velibs
    
    except Exception as e:
        print("Error occurred:", e)
        return velibs
```

```python
def export_data(df):
    try:
        timestamp = datetime.now().strftime("%Y-%m-%d_%H:%M")
        file_name = f'velibs_export_{timestamp}.csv'

        df.to_csv(file_name, index=False)
        print(f"Data exported to CSV file {file_name} successfully.")
    except Exception as e:
        print("Error occurred during data export:", e)
```

```python
iterations = 2
velibs = None

for i in range(iterations):
    velibs = collect_data(velibs)
    time.sleep(60)
```

```python
export_data(velibs)
```
